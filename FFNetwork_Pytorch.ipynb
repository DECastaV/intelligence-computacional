```python
import torch
import numpy as np
import matplotlib.pyplot as plt
import torchvision.transforms as transforms


from torch import nn
from torchvision.datasets import MNIST
from torch.utils.data import random_split, DataLoader
import torch.optim as optim
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

print("Using device:", device)
```

# Cargar y particionar el MNIST dataset

```python
mnist_train = MNIST(root = 'data/', train = True, transform = transforms.ToTensor(), download = True)
mnist_test = MNIST(root = 'data/', train = False, transform = transforms.ToTensor())
```

# Creación de clase de red neuronal + función de entrenamiento

```python
# Clase de Red Neuronal #1
class FeedForwardNN(nn.Module):
  def __init__(self, input_size, hidden_size, output_size):
    super().__init__()
    self.hidden = nn.Linear(input_size, hidden_size)
    self.output = nn.Linear(hidden_size, output_size)
    self.relu = nn.ReLU()

  def forward(self, x):
    x = x.reshape(-1, 784)
    x = self.hidden(x)
    x = self.relu(x)
    x = self.output(x)
    return x
```

```python
def calc_accuracy(y_pred, y_true):
  nan, y_prediction = torch.max(y_pred, 1)
  correct_predictions = torch.sum(y_prediction == y_true).item()
  accuracy = correct_predictions / len(y_pred)
  return accuracy
```

```python
def train(model, num_epochs, optimizer, batch_size):
  train_loader = DataLoader(mnist_train, batch_size = batch_size, shuffle = True)
  val_loader = DataLoader(mnist_train, batch_size = 10000, shuffle = True)

  loss_fn = nn.CrossEntropyLoss()

  for epoch in range(num_epochs):

    # Entrenar el modelo en batches
    for X_batch, y_batch in train_loader:
      y_pred = model(X_batch)
      loss = loss_fn(y_pred, y_batch)
      optimizer.zero_grad()
      loss.backward()
      optimizer.step()

    # Validación
    for X_i, y_i in val_loader:
      y_pred = model(X_i)
      val_loss = loss_fn(y_pred, y_i).item()
      val_accuracy = calc_accuracy(y_pred, y_i)

    print("Epoch: ", epoch+1, f"\tLoss: {val_loss:.4f} \tAccuracy: {val_accuracy:.4f}")
```

# Implementación del loop de entrenamiento
## Red Neuronal 1

- Capas ocultas: 120
- Épocas: 12
- Batches: 35
- Optimizer: SGD
- Learning Rate: 0.01

```python
# Entrenar el modelo con la clase establecida
model_m1 = FeedForwardNN(784, 120, 10)
optimizer_m1 = optim.SGD(model_m1.parameters(), lr = 0.01)
epochs_m1 = 12
batches_m1 = 35

model_m1.train()
train(model_m1, epochs_m1, optimizer_m1, batches_m1)
```

# Red Neuronal 2

- Capas ocultas: 240
- Épocas: 10
- Batches: 28
- Optimizer: Adam
- Learning Rate: 0.0075

```python
# Mismo modelo, misma clase, diferentes parámetros
model_m2 = FeedForwardNN(784, 240, 10)
optimizer_m2 = optim.Adam(model_m2.parameters(), lr = 0.0075)
epochs_m2 = 10
batches_m2 = 28


model_m2.train()
train(model_m2, epochs_m2, optimizer_m2, batches_m2)
```

```python
# Clase de Red Neuronal #2, con capas ocultas adicionales
class FeedForwardNN_ALT(nn.Module):
  def __init__(self, input_size, hidden_size, hidden_size_dos, hidden_size_tres, output_size):
    super().__init__()
    self.hidden = nn.Linear(input_size, hidden_size)
    self.hidden2 = nn.Linear(hidden_size, hidden_size_dos)
    self.hidden3 = nn.Linear(hidden_size_dos, hidden_size_tres)
    self.output = nn.Linear(hidden_size_tres, output_size)
    self.relu = nn.ReLU()

  def forward(self, x):
    x = x.reshape(-1, 784)
    x = self.hidden(x)
    x = self.relu(x)
    x = self.hidden2(x)
    x = self.relu(x)
    x = self.hidden3(x)
    x = self.relu(x)
    x = self.output(x)
    return x
```

# Red Neuronal 3

- Capa oculta 1: 100
- Capa oculta 2: 185
- Capa oculta 3: 190
- Épocas: 9
- Batches: 20
- Optimizer: SGD
- Learning Rate: 0.001

```python
# Clase diferente
model_m3 = FeedForwardNN_ALT(784, 100, 185, 190, 10)
optimizer_m3 = optim.SGD(model_m3.parameters(), lr = 0.001)
epochs_m3 = 9
batches_m3 = 20

model_m3.train()
train(model_m3, epochs_m3, optimizer_m3, batches_m3)
```

# Mejor modelo obtenido

Ganó la Red Neuronal 2, con un Loss de 0.0669 y un Accuracy de 0.9809.
Mientras que la Red Neuronal 1 obtuvo un Loss de 0.1695 y un Accuracy de 0.9535; y la Red Neuronal 3 obtuvo un Loss de 0.4075 y un Accuracy de 0.8804.
A continuación se evaluará este modelo con valores de prueba.

```python
test_data = DataLoader(mnist_test, batch_size=10000, shuffle=False)
loss_fn = nn.CrossEntropyLoss()

model_m2.eval()

# Evaluación con datos de prueba
with torch.no_grad():

  for X_test, y_test in test_data:

    y_pred = model_m2(X_test)
    loss = loss_fn(y_pred, y_test).item()
    acc = calc_accuracy(y_pred, y_test)


print(f"Test Loss: {loss:.4f}\n")
print(f"Accuracy (Global): {acc:.4f}\n")
```
